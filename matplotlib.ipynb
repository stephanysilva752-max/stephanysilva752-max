```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(
{'dias' : ['1', '4', '8', '12', '15'],#para aparecer as datas certinhas tem que transformar em str
'Venda' : [500, 200, 1000, 150, 1950]})

#PILOTANDO O GRFICO
plt.figure(figsize=(15,7))
plt.plot(df['dias'], df['Venda'], marker='o', linestyle='-', color='blue')
plt.title('Vendas por Dias')
plt.xlabel('Dia da Semana')
plt.ylabel('Valor Vendido')
plt.show()
```

```python
df = pd.DataFrame(
{'dias' : ['Segunda', 'Tera', 'Quarta', 'Quinta', 'Sexta'],
'Venda' : [500, 200, 1000, 150, 1950]})

#PILOTANDO O GRFICO
plt.barh(df['dias'], df['Venda'], color='blue')
plt.title('Vendas por Dias')
plt.xlabel('Dia da Semana')
plt.ylabel('Valor Vendido')
plt.show()
```
